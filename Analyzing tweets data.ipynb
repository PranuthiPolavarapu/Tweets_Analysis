```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Lets read the two csv files
user=pd.read_csv('userdata.csv')
tweet=pd.read_csv('tweetdata.csv')
```

```python
user.head(5)
```

```python
tweet.head(5)
```

```python
#Using the merge function and the outer join lets join two csv files on userId
data=pd.merge(user,tweet,on='userId',how='outer')
```

```python
data
```

```python
#check for null values
data.isnull()
```

```python
#drop the null values
cleaned=data.dropna(axis=1,how='all')
```

```python
cleaned.isnull()
```

```python
cleaned.fillna(0,inplace=True)
```

```python
cleaned
```

```python
df=pd.pivot_table(cleaned, values='followersCount',index='username',columns='verified')
```

```python
verified=df.sort_values(by=1,ascending=False)
```

```python
# Top 3 verified users based on the follower count
topver=verified.head(3)
topver
```

```python
chart1=topver.plot.bar(figsize=(10,8))
chart1.set_title('Top 3 verified users based on followers')
chart1.set_xlabel('username')
chart1.set_ylabel('followers')
```

```python
# top 5 users with more friends count
df1=pd.pivot_table(cleaned, values='friendsCount',index='username')
```

# 

```python
df1=df1.sort_values(by='friendsCount',ascending=False)
```

```python
df2=pd.pivot_table(cleaned, values='retweetCount',index='username')
```

```python
df2=df2.sort_values(by='retweetCount',ascending=False)
```

```python
topretweet=df2.head(5)
```
