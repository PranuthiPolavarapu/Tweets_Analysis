```python
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import warnings
```

```python
#read the json file
data=pd.read_json('farmers.json', lines=True)
```

```python
#to remove the data which is not in english
data=data[data['lang']=='en']
```

```python
data.head(10)
```

```python
data['user']
```

```python
#user column is in nested json format we can normalize it for better analysis
userdata=json_normalize(data['user'])
userdata.drop(['description', 'linkTcourl'], axis=1, inplace=True)
userdata.rename(columns={'id':'userId', 'url':'profileUrl'}, inplace=True)
```

```python
userdata.head(10)
```

```python
#We can add the normalized data into a dataframe for further processing
userdf=pd.DataFrame(userdata)
```

```python
userdf.head(10)
```

```python
# the users may have duplicate values lets remove duplicate values
userdf.drop_duplicates(subset=['userId'], inplace=True)
```

```python
userdf
```

```python
#Lets save the dataframe to csv format
userdf.to_csv('userdata.csv')
```

```python
#from the original data lets transform further
#add userid column so that we can join our user df for analysis
user_id = []
for user in data['user']:
    uid = user['id']
    user_id.append(uid)
data['userId'] = user_id
```

```python
# remove the unnecessary columns
cols = ['url', 'date', 'renderedContent', 'id', 'userId', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount', 'source', 'media', 'retweetedTweet', 'quotedTweet', 'mentionedUsers']
```

```python
tweet=data[cols]
```

```python
tweet.rename(columns={'id':'tweetId', 'url':'tweetUrl'}, inplace=True)
```

```python
tweet
```

```python
# lets create df for this data
tweetdata=pd.DataFrame(tweet)
```

```python
#Drop the duplicate data
tweetdata.drop_duplicates(subset=['tweetId'], inplace=True)
```

```python
tweetdata
```

```python
#lets convert it into csv file
tweetdata.to_csv('tweetdata.csv')
```
